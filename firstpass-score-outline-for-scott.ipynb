```python
# Step 11; No band, no date: score = 0
mask_11 = (pandas.isna(filenames_df['score'])) & (filenames_df['date_format'] == 'nodate') & (pandas.isna(filenames_df['abbrev_type']))
masked_filenames_11 = filenames_df[mask_11]
masked_filenames_11['score'].fillna(0, inplace=True)
filenames_df.update(masked_filenames_11)
filenames_df.head()
```

```python
# Step 3a: canonical abbreviation, date right after band, yyyy-mm-dd date: score = 4
mask3a = (pandas.isna(filenames_df['score'])) & (filenames_df['abbrev_type'] == 'canonical') & (filenames_df['date_format'] == 'yyyy-mm-dd') & (filenames_df['band_abbrev_after_date'] == False)
masked_filenames3a = filenames_df[mask3a]
masked_filenames3a['score'].fillna(4, inplace=True)
filenames_df.update(masked_filenames3a)
filenames_df.head()
```
